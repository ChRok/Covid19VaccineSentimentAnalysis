```python
#micro-batch generator that accepts the monthly data and “spits out” micro-batches of tweets with the number of tweets in a batch being completely random

from random import randrange

class MicroBatchGenerator:
    def __init__(self, list_data):
        self.list_data = list_data
        self.splitedData = self.splitDataRandomly()
        
    def splitDataRandomly(self):
        n = randrange(len(self.list_data))
#         print(n)
        splitedData = [self.list_data[i::n] for i in range(n)]
        return splitedData
```

```python
import numpy as np

class BitVector:
    def __init__(self, capacity=1):
        self.capacity = capacity
        self.vector = np.zeros((capacity,), dtype=int)

    def get(self, idx):
        return self.vector[idx]
        
    def add(self, idx):
        self.vector[idx] += 1

    def printVector(self):
        print("printVector: {}".format(self.vector))
```

```python
class CountMinSketch:
    def __init__(self, capacity, hashfuncs):
#         if (capacity > 1000):
#             raise ValueError("capacity must be less than or equal to 1000.")
        self.bitVector = BitVector(capacity)
        self.hashfuncs = hashfuncs

    def vec_hash_positions(self, elem):
        hash_posistions = []
        for i in range(self.hashfuncs):
#             print(i)
            hash_position = (hash(elem) + i) % self.bitVector.capacity
#             print(i)
#             print(hash(elem))
#             print((hash(elem) + i) % self.bitVector.capacity)
            hash_posistions.append(hash_position)
        return hash_posistions
    
    def increment(self, elem):
        hash_posistions = self.vec_hash_positions(elem)
        for idx in hash_posistions:
            self.bitVector.add(idx)
            
    def estimate(self, elem):
        hash_posistions = self.vec_hash_positions(elem)
#         print(hash_posistions)
        minCount = - 1
        for idx in hash_posistions:
            if (minCount == -1):
                minCount = self.bitVector.get(idx)
            else:
                if (minCount > self.bitVector.get(idx)):
                    minCount = self.bitVector.get(idx)
        return minCount 
```

```python
import pandas as pd
from textblob import TextBlob
import re

class EstimateMonthlySentimentAnalysis:
    def __init__(self, capacity, hashfuncs):
        self.countMinSketch = CountMinSketch(capacity, hashfuncs)
        self.hashfuncs = hashfuncs
        self.countMinSketch_sentiments = {'neutral': 0, 'positive': 0, 'negative': 0}
        self.actual_count_sentiments = {'neutral': 0, 'positive': 0, 'negative': 0}
        self.tweets = None
        
    def getSentimentAnalysis(self, tweets):
        self.tweets = tweets
        microBatchGenerator = MicroBatchGenerator(self.tweets)
        microBatces = microBatchGenerator.splitedData

        self.countMinSketch_sentiments = self.countSentimentsFrequency(microBatces)  

        return self.countMinSketch_sentiments     
        
    def countSentimentsFrequency(self, microBatces):
        day_sentiments = set()
        count_sentiments = {'neutral': 0, 'positive': 0, 'negative': 0}
        minSketch_count_sentiments = {'neutral': 0, 'positive': 0, 'negative': 0}
        for batch in microBatces:
#             print(batch)
            tweets = batch #['text']
#                 print(tweets)
            for tweet in tweets:
#                 print(tweet)
                sentiment = self.get_tweet_sentiment(tweet)
#                 print(sentiment)
                self.countMinSketch.increment(sentiment)
                day_sentiments.add(sentiment)
                count_sentiments[sentiment] += 1
        
        self.actual_count_sentiments = count_sentiments
        
        for sentiment in day_sentiments:
            sentiment_counter = self.countMinSketch.estimate(sentiment)
            minSketch_count_sentiments[sentiment] = sentiment_counter
        
        return minSketch_count_sentiments
    
    
    def get_tweet_sentiment(self, tweet):
        '''
        Utility function to classify sentiment of passed tweet
        using textblob's sentiment method
        '''
        # create TextBlob object of passed tweet text
        analysis = TextBlob(self.clean_tweet(tweet))
    #     print(analysis)
        # set sentiment
        if analysis.sentiment.polarity > 0:
            return 'positive'
        elif analysis.sentiment.polarity == 0:
            return 'neutral'
        else:
            return 'negative'
    
    def clean_tweet(self, tweet):
            '''
            Utility function to clean tweet text by removing links, special characters
            using simple regex statements.
            '''
            if not isinstance(tweet, str):
                return ""
            return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
    
    
    def clear(self):
        self.countMinSketch_sentiments = {}
        self.actual_count_sentiments = {}
        self.data = None
```

```python
# unpacking our sentiment_analysis

def unpack(df, column, fillna=None):
    ret = None
    if fillna is None:
        ret = pd.concat([df, pd.DataFrame((d for idx, d in df[column].iteritems()))], axis=1)
        del ret[column]
    else:
        ret = pd.concat([df, pd.DataFrame((d for idx, d in df[column].iteritems())).fillna(fillna)], axis=1)
        del ret[column]
    return ret

final_df = unpack(new_df, 'sentiment_analysis', 0)
final_df
```
